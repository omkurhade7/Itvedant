Id column was not that helpful and unnamed 32 column had all nan values in it so it has been dropped

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
df.head()
```

```python
#This step splits the data so we can train the data and then test the accuracy of it
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0],test_size=0.2, random_state=2)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
#Standardization to get all the columns on the same scale this helps to draw better corelation between the features and aslo
#one feature doesnot dominates the seen
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_train.shape
```

```python
#Importing the main class
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

```python
#Accuracy score gives us a percentage of how much correct predctions have the model given
from sklearn.metrics import accuracy_score

y_pred = knn.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
#This step is to see if the n values is changed can the accuracy be improved
#Here we are running a loop every time the loop is completed the accuracy score is stored in the list
scores = []

for i in range(1,16):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    
    knn.fit(X_train,y_train)
    
    y_pred = knn.predict(X_test)

    scores.append(accuracy_score(y_test, y_pred))
    
```

```python
#Visualizing this gives a better understanding of the accuracy
import matplotlib.pyplot as plt
plt.plot(range(1,16),scores)
```
