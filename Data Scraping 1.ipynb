```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
#These are the libraries that are needed for webscraping
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav',headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
#The BeautifulSoup object is made to scrape the data
```

```python
#Using the html tags to scrape the data
comp = soup.find_all('div',class_="companyCardWrapper")
#Now we have details of all the companies
```

```python
name = []
ratings = []
for i in comp:
    name.append(i.find('h2').text.strip())
    ratings.append(i.find('span',class_='companyCardWrapper__companyRatingValue').text.strip())
```

```python
reviews = []
salaries =[]
interviews =[]
jobs = []
benefits = []
```

```python
for i in comp:
    for j in range(5):
        if j == 0:
            reviews.append(i.find_all('a',class_='companyCardWrapper__ActionWrapper')[j].text.strip())
        elif j == 1:
            salaries.append(i.find_all('a',class_='companyCardWrapper__ActionWrapper')[j].text.strip())
        elif j == 2:
            interviews.append(i.find_all('a',class_='companyCardWrapper__ActionWrapper')[j].text.strip())
        elif j == 3:
            jobs.append(i.find_all('a',class_='companyCardWrapper__ActionWrapper')[j].text.strip())
        elif j == 4:
            benefits.append(i.find_all('a',class_='companyCardWrapper__ActionWrapper')[j].text.strip())
```

```python
comp_type = []
emp_count =[]
comp_type2 =[]
comp_age =[]
```

```python
for i in comp:
    comp_type.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[0])
    emp_count.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[1])
    comp_type2.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[2])
    comp_age.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[3])
```

```python
comp1 = {
    'name':name,
    'ratings':ratings,
    'reviews':reviews,
    'salaries':salaries,
    'jobs':jobs,
    'benefits':benefits,
    'comp_type':comp_type,
    'emp_count':emp_count,
    'comp_type2':comp_type2,
    'comp_age':comp_age
}
```

```python
companies = pd.DataFrame(comp1)
```
